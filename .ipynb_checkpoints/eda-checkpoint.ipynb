```python
from allensdk.core.cell_types_cache import CellTypesCache
import pandas as pd
```

```python
ctc = CellTypesCache(manifest_file="ctc.json")
cells = ctc.get_cells()
cells_df = pd.DataFrame(cells)
cells_df.set_index('id', inplace=True)
cells_df
```

```python
morph = ctc.get_morphology_features(dataframe=True)
morph.set_index('specimen_id', inplace=True)
morph.head()
```

```python
ephys = ctc.get_ephys_features(dataframe=True)
ephys.set_index('specimen_id', inplace=True)
ephys.head()
```

```python
ephys_cells = cells_df.join(ephys)
ephys_cells.columns
```

```python
# avoid overlapping column names (e.g., 'id') by adding a suffix for morph columns
df = ephys_cells.join(morph, rsuffix='_morph')
df.columns
```
