```python
# import allensdk and set up CellTypesCache
from allensdk.core.cell_types_cache import CellTypesCache
from allensdk.api.queries.cell_types_api import CellTypesApi
ctc = CellTypesCache(manifest_file="cell_types/manifest.json")
```

```python
# import numpy and pandas 
import numpy as np
import pandas as pd
```

```python
# get all cells
all_cells = ctc.get_cells()
# turn to dataframe for easier visualization
all_cells_df = pd.DataFrame(all_cells)
all_cells_df.head()
```

```python
# view all columns
all_cells_df.columns.tolist()
```

```python
# feeatures for electrophysiology
ephys_features = ctc.get_ephys_features()
ephys_features_df = pd.DataFrame(ephys_features)
ephys_features_df.head()
```
